```python
fig, ax = plt.subplots(figsize=(15,15))
ax.scatter(x=df['gdp_per_capita'],y=df['total_deaths_per_million'],c='White')
ax.set_xlabel('GDP per Capita',fontdict={'fontsize':24})
ax.set_ylabel('Total Deaths per Million',fontdict={'fontsize':24})
for idx, row in df.iterrows():
    ax.annotate(row['iso_code'],(row['gdp_per_capita'],row['total_deaths_per_million']))
plt.show()
```

```python
import statsmodels.api as sm
import numpy as np
```

```python
# There were 'NA' entries that prevented the analysis, had to drop them!
df4 = df.dropna(subset=['median_age', 'total_deaths_per_million'])
df4.shape
```

```python
X = df4[['median_age']]
y = df4['total_deaths_per_million']
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X).fit()
predictions = model.predict(X)
model.summary()
```
